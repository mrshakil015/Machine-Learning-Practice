# K-Nearest Neighbors(K-NN) Classifier

#### Import the Libriras

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

#### Load the Dataset

```python
data = pd.read_csv('User_Data.csv')
data
```

```python
data.shape
```

```python
x= data.iloc[:,2:4].values
x
```

```python
x.shape
```

```python
y= data.iloc[:,-1]
y
```

#### Splitting the dataset into training and test set

```python
from sklearn.model_selection import train_test_split
Xtrain, Xtest, Ytrain, Ytest = train_test_split(x,y,test_size=0.20)
```

```python
Xtrain.shape
```

```python
Xtest.shape
```

#### Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
Xtrain = scaler.fit_transform(Xtrain)
Xtrain
```

```python
Xtest = scaler.fit_transform(Xtest)
Xtest
```

#### Claculate the value of K

```python
k = round(np.sqrt(Xtrain.shape[0]))
k1 = k%2
if k1==0:
    k=k-1;
else:
    k=k;
k
```

#### Fitting the K-NN Classifier

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=k)
knn.fit(Xtrain,Ytrain)
```

#### Predict the result

```python
Ypred= knn.predict(Xtest)
Ypred
```

#### Confiusion Matrix

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(Ytest,Ypred)
```

#### Accuracy Score

```python
from sklearn.metrics import accuracy_score
print("Decision Tree Accuracy Score for Test dataset: ",accuracy_score(Ytest,Ypred)*100,"%")
```

```python
print("Decision Tree Accuracy Score for Train dataset: ",knn.score(Xtrain,Ytrain)*100,"%")
```

#### Random Accuracy Check

```python
accuracy=[]
for i in range(1,26):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(Xtrain,Ytrain)
    accuracy.append(accuracy_score(Ytest,knn.predict(Xtest)))
```
